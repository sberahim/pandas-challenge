```python
# Dependencies and Setup.
import pandas as pd

# File to Load.
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames.
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Find the total number of schools.
totalNoofSchools = school_data['school_name'].count()

# Find the total number of students.
totalNoofStudents = school_data_complete['student_name'].count()

# Get the total school budgets.
totalBudgets = sum(school_data_complete['budget'].unique())

# Get the average total maths scores and round to 6 decimal points.
avgMathScore = school_data_complete['maths_score'].mean().round(6)

# Get the average total reading score and round to 6 decimal points.
avgReadingScore = school_data_complete['reading_score'].mean().round(6)

# Calculate the total passing maths score percentage. Passing marks is 50 and above. Final value is round to 6 decimal points.
passMathScorePercent = (((school_data_complete[school_data_complete['maths_score'] > 49]['maths_score'].count()) / totalNoofStudents) * 100).round(6)

# Calculate the total passing reading score percentage. Passing marks is 50 and above. Final value is round to 6 decimal points.
passReadingScorePercent = (((school_data_complete[school_data_complete['reading_score'] > 49]['reading_score'].count()) / totalNoofStudents) * 100).round(6)

# Calculate the overall passing percentage for student that has passed both maths and reading. Passing marks is 50 and above. Final value is round to 6 decimal points.
overallPassMRPercent = ((len(school_data_complete[(school_data_complete['reading_score'] > 49) & (school_data_complete['maths_score'] > 49)]) / totalNoofStudents) * 100).round(6)

# Store each value into dataframe.
lgaSummary = pd.DataFrame({"Total Schools":totalNoofSchools,
                           "Total Students":totalNoofStudents,
                           "Total Budget":totalBudgets,
                           "Average Math Score":avgMathScore,
                           "Average Reading Score":avgReadingScore,
                           "% Passing Maths":passMathScorePercent,
                           "% Passing Reading":passReadingScorePercent,
                           "% Overall Passing":overallPassMRPercent}, index=[0])

# Display the results with correct format.
lgaSummary.style.format({"Total Students": '{:,}',"Total Budget": '${:,.2f}'})
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Remove any duplicates for school name and group damn by alphabet order.
schoolName = school_data_complete.groupby('school_name')

# Get the school type from school data set.
schoolType = school_data.set_index('school_name')['type']

# Get the total number of students for each school.
totalStudents = schoolName['student_name'].count()

# Get each school budget from the school data set.
eachSchoolBudget = school_data.set_index('school_name')['budget']

# Calculate budget spent for each student per school.
studentBudget = eachSchoolBudget / totalStudents

# Get the average maths score for each school and round to 6 decimal points.
avgMathScorePerSch =schoolName['maths_score'].mean().round(6)

# Get the average reading score for each school and round to 6 decimal points
avgReadScorePerSch =schoolName['reading_score'].mean().round(6)

# Calculate the total passing maths score percentage per school. Passing marks is 50 and above. Final value is round to 6 decimal points.
passMathScorePerSch = (((school_data_complete[school_data_complete['maths_score'] > 49].groupby('school_name')['student_name'].count()) / totalStudents) * 100).round(6)

# Calculate the total passing reading score percentage per school. Passing marks is 50 and above. Final value is round to 6 decimal points.
passReadScorePerSch = (((school_data_complete[school_data_complete['reading_score'] > 49].groupby('school_name')['student_name'].count()) / totalStudents) * 100).round(6)

# Calculate the overall passing percentage for student that has passed both maths and reading for each school. Passing marks is 50 and above. Final value is round to 6 decimal points.
overallPassPerSch = (((school_data_complete[(school_data_complete['reading_score'] > 49) & (school_data_complete['maths_score'] > 49)].groupby('school_name')['student_name'].count()) / totalStudents) * 100).round(6)

# Store each value into dataframe.
per_school_summary = pd.DataFrame({
   "School Type":schoolType,
   "Total Students":totalStudents,
   "Total School Budget":eachSchoolBudget,
   "Per Student Budget":studentBudget,
   "Average Maths Score":avgMathScorePerSch,
   "Average Reading Score":avgReadScorePerSch,
   "% Passing Maths":passMathScorePerSch,
   "% Passing Reading":passReadScorePerSch,
    "% Overall Passing":overallPassPerSch
})

# Remove the column name for school.
per_school_summary.index.name = None

# Display the results with correct format.
per_school_summary.style.format({"Total School Budget": "${:,.2f}","Per Student Budget": "${:.2f}"})
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Get the results from the previous dataframe and store it to a new data frame called top_schools. 
# Sort the value by overall passing pacentage in decending order.
top_schools = per_school_summary.sort_values("% Overall Passing", ascending = False)

# Display the top 5 results with correct format.
top_schools.head().style.format({"Total School Budget": "${:,.2f}","Per Student Budget": "${:.2f}"})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Declare the bins and labels.
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Store the results from previous school summary dataframe to a new dataframe called school_spending_df.
school_spending_df = per_school_summary

# Categorise data set based on each student spending range.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Total School Budget"]/school_spending_df["Total Students"], spending_bins, labels = labels)

# Get the value of average maths score, reading score, passing maths score percentage, passing reading score percentage and overall passing score percentage
# based on each student spending range.
spending_maths_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

# Store values into a new dataframe called spending_summary.
spending_summary = pd.DataFrame({
   "Average Maths Score":spending_maths_scores,
   "Average Reading Score":spending_reading_scores,
   "% Passing Maths":spending_passing_maths,
   "% Passing Reading":spending_passing_reading,
   "% Overall Passing":overall_passing_spending
})

#Display results with correct format.
spending_summary.style.format({"Average Maths Score": "{:,.2f}",
                               "Average Reading Score": "{:.2f}",
                               "% Passing Maths": "{:.2f}",
                               "% Passing Reading": "{:.2f}",
                               "% Overall Passing": "{:.2f}",})
```
